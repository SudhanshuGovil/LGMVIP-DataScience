### Importing CV2 library

```python
import cv2
import matplotlib.pyplot as plt
```

### Read image

```python
img=cv2.imread("image.jpg")
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.axis(False)
plt.show()
```

### Converting Image to Pencil sketch

#### Convert to Grey Image

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_img)
plt.axis(False)
plt.show()
```

#### Invert Image

```python
invert_img=255-grey_img
```

```python
plt.imshow(invert_img)
plt.axis(False)
plt.show()
```

#### Blur Image

```python
blur_img=cv2.GaussianBlur(invert_img, (99,99),0)
```

```python
plt.imshow(blur_img)
plt.axis(False)
plt.show()
```

#### Invert Blurred Image

```python
invblur_img=255-blur_img
```

```python
plt.imshow(invblur_img)
plt.axis(False)
plt.show()
```

#### Sketch Image

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
```

```python
plt.imshow(sketch_img)
plt.axis(False)
plt.show()
```

```python
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_sketch)
plt.axis(False)
plt.show()
```

### Original Vs Sketch Image

```python
plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow(rgb_img)
plt.axis(False)
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(rgb_sketch)
plt.axis(False)
plt.title('Sketch')

plt.show()
```
